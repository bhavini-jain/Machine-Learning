```python
sns.barplot(x='size',y='total_bill',hue='sex',data=tips_df,estimator=np.max,palette='magma',dodge=False,**k)
```

```python
sns.barplot(x='size',y='total_bill',hue='sex',data=tips_df,estimator=np.max,palette='hot',dodge=False,alpha= 0.8, linestyle= '-.', linewidth= 2, edgecolor= 'k')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,9))
ax=sns.barplot(x='size',y='total_bill',hue='sex',data=tips_df,estimator=np.max,palette='hot',dodge=False,alpha= 0.8, linestyle= '-.', linewidth= 2, edgecolor= 'k')

ax.set(title='Size Vs Total Bill',xlabel='Size',ylabel='Total Bill')
plt.savefig("figure")
plt.show()
```
